```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf


data = pd.read_csv("acn_data.csv")

# Extract the energy stored values from the dataset
energy_stored = data['ACN data']

# Create a time series for x-axis (assuming one data point per time unit)
time_series = range(len(energy_stored))
plt.rcParams['font.size'] = '14'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (8, 6)
# Plot the energy stored over time
plt.plot(time_series, energy_stored, label='Temporal variation')
plt.xlabel('Days')
plt.title('ACN data')
plt.legend()
plt.grid()
plt.savefig('ACN_data_description.eps',bbox_inches='tight') 
plt.savefig('ACN_data_description.pdf',bbox_inches='tight') 
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("acn_data.csv")

# Create a custom-styled box plot
plt.rcParams['font.size'] = '14'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (8, 6)

# Create a box plot with custom styles
boxplot = data.boxplot(column=['ACN data', 'Palo Alto data', 'Perth data'], 
                       vert=True,  # Horizontal orientation
                       patch_artist=True,  # Filled boxes
                       medianprops={'color': 'royalblue'},  # Median line color
                       boxprops={'facecolor': 'darkseagreen', 'edgecolor': 'darkseagreen'},  # Box properties
                       whiskerprops={'color': 'purple'},  # Whisker color
                       capprops={'color': 'red'},  # Cap color
                       flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markeredgecolor': 'orange'}  # Outlier style
                      )



plt.savefig('box_plot_1.eps',bbox_inches='tight') 
plt.savefig('box_plot_1.pdf',bbox_inches='tight') 
plt.show()
```
