# ACN DATA GRAPHICALL REPRESENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("ACN_OUTPUT.csv")
features=list(df)
df= df.head(60)
print(features)
plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (8, 6)
plt.plot(df['Actual'], label= 'Measured',marker = ".")
plt.plot(df['Hybrid'],label='DeepBoost' ,marker = '+')
#plt.plot(df['LSTM'],label='LSTM',marker = 's' )
#plt.plot(df['CatBoost'],label='CatBoost',marker = '*')
#plt.plot(df['XgBoost'],label='XgBoost',marker = "o")
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel('Days', fontweight="bold")
plt.title("ACN charging stattion", fontweight="bold")
plt.ylabel('EVs charging load (kWh)', fontweight="bold")
plt.legend(prop={"size":16})
plt.savefig('ACN_OUTPUT.pdf',bbox_inches='tight')
plt.savefig('ACN_OUTPUT.eps',bbox_inches='tight')
plt.savefig('ACN_OUTPUT.png',bbox_inches='tight')
plt.savefig('ACN_OUTPUT.svg',bbox_inches='tight')
```

# Palo Alto GRAPHICALL REPRESENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("Palo Alto_output.csv")
features=list(df)
df= df.head(60)
print(features)
plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (8, 6)
plt.plot(df['Actual'], label= 'Measured',marker = ".",)
plt.plot(df['Hybrid'],label='DeepBoost' ,marker = '+',)

plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel('Days',fontweight='bold')
plt.ylabel('EVs charging load', fontweight="bold")
plt.title("Palo Alto charging station (kWh)", fontweight="bold")
plt.legend(prop={"size":16})
plt.savefig('PALO ALTO_OUTPUT.pdf',bbox_inches='tight')
plt.savefig("PALO ALTO_OUTPUT.eps",bbox_inches='tight')
plt.savefig('PALO ALTO_OUTPUT.png',bbox_inches='tight')
plt.savefig('PALO ALTO_OUTPUT.svg',bbox_inches='tight')
```

# PERTH DATASET GRAPHICALL REPRESENTATION
