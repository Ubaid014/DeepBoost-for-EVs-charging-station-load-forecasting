# RMSE BAR PLOT

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
d = {'Hybrid': [2.61,7.63,13.83], 'CatBoost': [3.01,13,27.45],'XgBoost': [4.09,13.37,27.36],'LSTM': [18.58,30.14,32.36]}
data=pd.DataFrame(data=d, index=['ACN', 'Palo Alto', 'Perth'])
print(data)
x1=data['Hybrid']
x2=data['CatBoost']
x3=data['XgBoost']
x4=data['LSTM']
y=["ACN","Palo Alto","Perth"]
sns.set_style("white")
y_axis=np.arange(len(y))
plt.ylabel("RMSE (kWh)",fontweight="bold")
plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(prop={"size":16})
hatches = ['*', '.', 'x', '/']
plt.rcParams["figure.figsize"] = (8,6)
plt.bar(y_axis, x1,0.2, label= 'DeepBoost' ,edgecolor='black', color=['silver'],hatch=hatches[0])
plt.bar(y_axis+0.2, x2,0.2, label= 'CatBoost' ,edgecolor='black', color=['royalblue'],hatch=hatches[1])
plt.bar(y_axis+0.4, x3,0.2, label= 'XgBoost',edgecolor='black',color=['sandybrown'],hatch=hatches[2])
plt.bar(y_axis+0.6, x4,0.2, label= 'LSTM',edgecolor='black', color=['darkseagreen'],hatch=hatches[3])
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.legend()
plt.savefig('RMSE_barplot.pdf') 
plt.savefig('RMSE_barplot.eps') 
plt.savefig('RMSE_barplot.png') 
plt.savefig('RMSE_barplot.svg') 

```
